```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
height = 8
width  = 8
image = np.ones((height,width,3), dtype="uint8")

image[:,4:6] = 255
image[0:2,2:4] = 255
image[3:5,2:4] = 255
image[6:8,2:4] = 255


plt.title('Color image in RGB format')
plt.imshow(image)
plt.show()
```

```python
height = 8
width  = 8
image = np.ones((height,width,3), dtype="uint8")

image*= 255

# Red component
image[:,4:6,0] = 255
image[0:2,2:4,0] = 255
image[3:5,2:4,0] = 255
image[6:8,2:4,0] = 255

# Green component
image[:,4:6,1] = 0
image[0:2,2:4,1] = 0
image[3:5,2:4,1] = 0
image[6:8,2:4,1] = 0

# Blue component
image[:,4:6,2] = 0
image[0:2,2:4,2] = 0
image[3:5,2:4,2] = 0
image[6:8,2:4,2] = 0


plt.title('Color image in RGB format')
plt.imshow(image)
plt.show()
```
