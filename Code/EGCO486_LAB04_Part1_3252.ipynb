Exercise1: Pseudo-color image processing using multi-level intensity slicing

```python
image1_path = path + 'MRI.bmp'

image1 = cv2.imread(image1_path,0)

output = np.zeros([image1.shape[0],image1.shape[1],3],np.uint8)

for i in range(image1.shape[0]):
  for j in range(image1.shape[1]):

    #BLUE
    if image1[i,j] >= 0 and image1[i,j] <= 30:
      output[i,j] = (0,0,255)
    #YELLOW
    elif image1[i,j] >= 100 and image1[i,j] <= 140:
      output[i,j] = (255,255,0)
    else:
      output[i,j] = image1[i,j]
plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.title('original')
plt.imshow(image1,cmap="gray")

plt.subplot(1, 2, 2)
plt.title('after slicing color')
plt.imshow(output)

plt.show()
```
