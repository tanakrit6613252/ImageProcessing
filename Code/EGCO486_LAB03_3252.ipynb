# **Exercise1: Image enhancement in the spatial domainusingimage negatives**

```python
image1_path = path + "breast_digital_Xray.tif"

def Myneg(image1_path):
  image1 = cv2.imread(image1_path, 0)
  image1_neg = 255 - image1

  plt.subplot(1,2,1)
  plt.title("Input Image")
  plt.imshow(image1, cmap='gray')
  plt.xticks([]), plt.yticks([])

  plt.subplot(1,2,2)
  plt.title("Output Image")
  plt.imshow(image1_neg, cmap='gray')
  plt.xticks([]), plt.yticks([])

  plt.show()

my_neg = Myneg(image1_path)
```

# **Exercise2: Image enhancement in the spatial domainusing power-law transformations**

```python
image2_path = path + "fractured_spine.tif"

def Mypow(image2_path):
  image2 = cv2.imread(image2_path, 0)

  plt.subplot(1,2,1)
  plt.title("Input Image")
  plt.imshow(image2, cmap='gray')
  plt.xticks([]), plt.yticks([])

  plt.subplot(1,2,2)
  plt.title("Output Image")
  image2_pow = np.power(image2, 0.4)
  plt.imshow(image2_pow, cmap='gray')
  plt.xticks([]), plt.yticks([])

  plt.show()

Mypow(image2_path)
```

# **Exercise3: Image enhancement in the spatial domainusing gray level slicing**

3.1

```python
image3_path = path + "MRI.bmp"

def Mygrayls(image3_path):
  image3 = cv2.imread(image3_path, 0)

  plt.subplot(1,2,1)
  plt.title("Input Image")
  plt.imshow(image3, cmap='gray')
  plt.xticks([]), plt.yticks([])


  for i in range(image3.shape[0]):
    for j in range(image3.shape[1]):
      if 100 <= image3[i][j] <= 140:
        image3[i][j] = 255
      else:
        image3[i][j] = 0

  plt.subplot(1,2,2)
  plt.title("Output Image")
  plt.xticks([]), plt.yticks([])
  plt.imshow(image3, cmap='gray')

  plt.show()

Mygrayls(image3_path)
```

# **Exercise4: Image enhancement in the spatial domainusing bit plane slicing**

```python
image4_path = path + "100-dollars.tif"

def Mybitps(image4_path):
  image4 = cv2.imread(image4_path, 0)

  plt.subplot(2,1,1)
  plt.title("Input Image")
  plt.imshow(image4, cmap='gray')
  plt.xticks([]), plt.yticks([])

  bit_planes = np.zeros((8, image4.shape[0], image4.shape[1]), dtype=np.uint8)

  for k in range(8):
    for i in range(image4.shape[0]):
      for j in range(image4.shape[1]):
        bit = image4[i][j] & (2**k)
        bit_planes[k][i][j] = bit

  bit_planes_7654 = bit_planes[4]

  plt.subplot(2,1,2)
  plt.title("Output Image")
  plt.imshow(bit_planes_7654, cmap='gray')
  plt.xticks([]), plt.yticks([])

  plt.show()


Mybitps(image4_path)
```
