```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
image1 = np.zeros([256,256], dtype="uint8")

for i in range(256):
  image1[:,i] = i


plt.title('Image of a ramp (256 gray levels)')
plt.imshow(image1, cmap='gray')
plt.show()
```

```python
image2 = cv2.imread(path + 'moon_dark.bmp', 0)

title = math.sqrt(image2.size)

for i in range(6):
  plt.subplot(2,3,i+1)
  plt.title(str(int(title/(2**i))) + ' x ' + str(int(title/(2**i))))
  plt.xticks([]), plt.yticks([])
  plt.imshow(image2[::2**i,::2**i], cmap='gray')

plt.show()
```
